```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import cross_val_score

#creting a pipeline with the Decision Tree model as its only step 
my_pipeline= make_pipeline(DecisionTreeClassifier(random_state=0))
#getting the accuracy of the model by cross valodation score with partition of data into 5 parts
cv_scores=cross_val_score(my_pipeline,
                         x,y,
                         cv=5,
                         scoring='accuracy')
print(cv_scores)
```

```python
cv_scores.mean()
```

```python
#function to calculate the cross valodation score for a given pipeline
def calculate_csvs(pipeline1):
    cv_scores=cross_val_score(pipeline1,
                         x,y,
                         cv=2,
                         scoring='accuracy')
    #print("CV score",cv_scores)
    return(cv_scores.mean())
```

```python
from sklearn.ensemble import RandomForestClassifier

#function to create a RandomForestClassifier for the data with number of trees used for estimation as the input
def rfr(n_estimator):
    #print("The value of n_estimators is", n_estimator)
    my_p2=make_pipeline(RandomForestClassifier(n_estimators=n_estimator))
    #call to calculate_csv function
    mean_csv=calculate_csvs(my_p2)
    return(mean_csv)
```

```python
rfr(200)
```

```python
#storing the results of various n_estimators as a decitonary where n_estimator:mean_csv
results={}
for i in range(1,20):
    results[i*50]=rfr(i*50)
    
#plotting the graph for various values of n_estimators    
plt.plot(list(results.keys()),list(results.values()))
plt.show
```

```python
#sorting the dictionary based on values
sorted_results={}
sorted_results=sorted(results.items(),key=lambda x : x[1],reverse=True)
for i in sorted_results:
    print(i[0] , " ", i[1])
```

```python
#maximum value od dictionary
max_n = max(results, key=results.get)
print("Print using max, n_estimator is ",max_n)

#another way could be by printing the key that position
print("Print using position indexing, n_estimator is  ",sorted_results[0][0],"\nAccuracy is ",sorted_results[0][1])
```
